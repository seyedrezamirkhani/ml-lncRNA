Let's Read *Attention Is All You Need*

```python
file_path = f'{LANDMARK_FILES_FOLDER}NIPS-2017-attention-is-all-you-need-Paper.pdf'

try:
    # Open the PDF file
    with open(file_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        text = ""
        
        # Extract text from each page
        for page in reader.pages:
            text += page.extract_text()
        
        # Detect the language of the extracted text
        if text.strip():  # Ensure there's text to analyze
            doc = nlp(text)
        else:
            print(f"No text found in {file_path}")
except Exception as e:
    print(f"Error processing {file_path}: {e}")
```

```python
print(doc.ents)
```

```python
options = {"colors": {"ML_METHOD": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}}
displacy.render(doc, style="ent", options=options, jupyter=True)
```

# Test robustness

```python
mod_text = text.replace('neural networks', 'neural netwerks')
```

```python
mod_doc = nlp(mod_text)
```

```python
print(mod_doc.ents)
```

**Excellent, the model identified the typo 'Neural Netwerks'**



**Compare this rule matching method**

```python
ruler_nlp = spacy.load("en_core_web_sm", exclude=["ner"])  # Exclude the default NER component
ruler = ruler_nlp.add_pipe("entity_ruler", last=True)

# Load the EntityRuler patterns from a file
ruler.from_disk(f"{DATA_FOLDER}ml_entity_ruler_patterns")
```

```python
ruler_mod_doc = ruler_nlp(mod_text)
```

```python
print(ruler_mod_doc.ents)
```

**We can see that the enity ruler is unable to match for typo 'Neural Netwerks' and the ML model we trained with spaCy can!**



We can visualise the matches using the EntityRuler and see how 'Neural Networks' is not identified

```python
options = {"colors": {"ML_METHOD": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}}
displacy.render(ruler_mod_doc, style="ent", options=options, jupyter=True)
```
